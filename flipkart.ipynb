```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings("ignore")
import time
```

```python
driver = webdriver.Chrome(r"C:\web driver\chromedriver.exe")
```

```python
driver=webdriver.Chrome("chromedriver.exe")
time.sleep(2)
```

```python
driver.get('https://www.flipkart.com/')
```

```python
search_for_products=driver.find_element_by_class_name("_3704LK")
search_for_products.send_keys("sunglasses")
```

```python
search_button=driver.find_element_by_xpath("/html/body/div[1]/div/div[1]/div[1]/div[2]/div[2]/form/div/button")
search_button.click()
```

```python
brand_titles = []
description_products = []
price_list = []
```

```python
titles_tags=driver.find_elements_by_xpath("//div[@class='_2WkVRV']")
titles_tags
```

```python
for i in titles_tags:
    title=i.text
    brand_titles.append(title)
brand_titles
```

```python
description_tags=driver.find_elements_by_xpath("//a[@class='IRpwTa']")
description_tags
```

```python
for i in description_tags:
    description_product=i.text
    description_products.append(description_product)
description_products
```

```python
price_tags=driver.find_elements_by_xpath("//div[@class='_30jeq3']")
price_tags
```

```python
for i in price_tags:
    price=i.text
    price_list.append(price)
price_list
```

```python
brands=pd.DataFrame({})
brands['title']=brand_titles
brands['description']=description_products
brands['price']=price_list
```
